### When to use a Line Chart?

Use a line chart when your data has a **natural order** — most commonly time (days, months, years). It's perfect for answering: *"How did X change over time?"*

**Real-world uses:** Stock prices, website traffic, temperature trends, sales growth

> **Key insight:** Multiple lines = comparisons across time. Don't use more than 5–6 lines or it becomes unreadable!

````python
# ============================================================
# LINE CHART — Monthly Revenue for 3 Products
# ============================================================
np.random.seed(42)
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
x = np.arange(len(months))

product_a = [42, 45, 48, 52, 55, 60, 65, 63, 70, 74, 80, 88]
product_b = [30, 32, 29, 35, 40, 38, 42, 50, 48, 55, 58, 62]
product_c = [20, 22, 25, 23, 28, 32, 30, 35, 40, 38, 45, 50]

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor('#fafbff')
ax.set_facecolor('#fafbff')

# Plot each line with markers
for data, label, color, marker in zip(
    [product_a, product_b, product_c],
    ['Product A', 'Product B', 'Product C'],
    ['#f72585', '#4361ee', '#06d6a0'],
    ['o', 's', '^']
):
    ax.plot(x, data, color=color, linewidth=2.5,
            marker=marker, markersize=7, label=label)
    # Shaded area under line
    ax.fill_between(x, data, alpha=0.08, color=color)

# Annotate the final values
for data, color in zip([product_a, product_b, product_c],
                        ['#f72585', '#4361ee', '#06d6a0']):
    ax.annotate(f'${data[-1]}K',
                xy=(11, data[-1]),
                xytext=(11.15, data[-1]),
                color=color, fontsize=10, fontweight='bold',
                va='center')

ax.set_xticks(x)
ax.set_xticklabels(months)
ax.set_ylabel('Revenue ($ thousands)')
ax.set_title('Monthly Revenue by Product — 2024', pad=15)
ax.legend(loc='upper left', framealpha=0.9, edgecolor='#dddddd')
ax.grid(axis='y', linestyle='--', alpha=0.4, color='#aaaaaa')
ax.set_xlim(-0.3, 12.5)

plt.tight_layout()
plt.show()

# TRY: Change the values in product_a/b/c to your own data!
# TRY: Add a 4th product line using the same pattern
````

### When to use a Bar Chart?

Bar charts are the **workhorse of data visualization** — use them whenever you're comparing values across **distinct categories**.

| Variant | When to Use |
|---------|------------|
| Vertical bars | Comparing categories (default choice) |
| Horizontal bars | When category names are long |
| Grouped bars | Comparing sub-groups side-by-side |
| Stacked bars | Showing part-to-whole within groups |

> **Pro tip:** Always sort your bars by value unless the order is meaningful (like months). Sorted bars are faster to read!

````python
# ============================================================
# BAR CHART — Grouped + Horizontal variants
# ============================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor('#fafbff')

# ── LEFT: Grouped Bar Chart
ax1.set_facecolor('#fafbff')
categories = ['Python', 'JavaScript', 'Java', 'C++', 'Rust']
jobs_2023  = [45, 38, 30, 22, 8]
jobs_2024  = [52, 40, 28, 20, 15]
x = np.arange(len(categories))
width = 0.35

bars1 = ax1.bar(x - width/2, jobs_2023, width, label='2023',
                color='#4361ee', alpha=0.85, edgecolor='white', linewidth=0.8)
bars2 = ax1.bar(x + width/2, jobs_2024, width, label='2024',
                color='#f72585', alpha=0.85, edgecolor='white', linewidth=0.8)

# Value labels on bars
for bar in list(bars1) + list(bars2):
    h = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, h + 0.5,
             f'{int(h)}K', ha='center', va='bottom',
             fontsize=8.5, fontweight='bold', color='#333')

ax1.set_xticks(x)
ax1.set_xticklabels(categories)
ax1.set_ylabel('Job Postings (thousands)')
ax1.set_title('Dev Job Postings by Language', pad=12)
ax1.legend(edgecolor='#dddddd')
ax1.grid(axis='y', linestyle='--', alpha=0.35, color='#aaaaaa')
ax1.set_ylim(0, 65)

# ── RIGHT: Horizontal Bar Chart (sorted)
ax2.set_facecolor('#fafbff')
frameworks = ['React', 'Django', 'FastAPI', 'Flask', 'Vue', 'Angular', 'Spring']
popularity = [72, 55, 48, 45, 40, 35, 30]

# Sort by value
sorted_pairs = sorted(zip(popularity, frameworks))
popularity_s, frameworks_s = zip(*sorted_pairs)

colors = [PALETTE[i % len(PALETTE)] for i in range(len(frameworks_s))]
bars = ax2.barh(frameworks_s, popularity_s, color=colors,
                alpha=0.88, edgecolor='white', linewidth=0.8, height=0.6)

for bar, val in zip(bars, popularity_s):
    ax2.text(val + 0.5, bar.get_y() + bar.get_height()/2,
             f'{val}%', va='center', fontsize=9,
             fontweight='bold', color='#333')

ax2.set_xlabel('Developer Popularity (%)')
ax2.set_title('Most Popular Frameworks (2024)', pad=12)
ax2.set_xlim(0, 85)
ax2.grid(axis='x', linestyle='--', alpha=0.35, color='#aaaaaa')

plt.tight_layout(pad=2)
plt.show()

# TRY: Change 'categories' and values to your own dataset
# TRY: Make a stacked bar with: ax.bar(x, data1) then ax.bar(x, data2, bottom=data1)
````

### When to use a Scatter Plot?

Scatter plots reveal **relationships (correlations)** between two numerical variables. Every dot = one data point.

You can encode **4 variables** in a single scatter plot:
- X position → variable 1
- Y position → variable 2  
- Dot **color** → variable 3 (category)
- Dot **size** → variable 4 (magnitude)

> This 4-variable scatter plot is called a **Bubble Chart** — one of the most information-dense charts available!

````python
# ============================================================
# SCATTER / BUBBLE CHART — Country Development Data
# ============================================================
np.random.seed(7)

# Simulated country data
regions = ['Asia', 'Europe', 'Americas', 'Africa']
region_colors = ['#f72585', '#4361ee', '#06d6a0', '#ffd166']

fig, ax = plt.subplots(figsize=(11, 6))
fig.patch.set_facecolor('#fafbff')
ax.set_facecolor('#fafbff')

for region, color in zip(regions, region_colors):
    n = 12
    gdp         = np.random.uniform(1000, 60000, n)
    life_exp    = 50 + (gdp / 60000) * 35 + np.random.normal(0, 3, n)
    population  = np.random.uniform(5, 200, n)

    scatter = ax.scatter(
        gdp, life_exp,
        s=population * 3,        # Size = population
        c=color,
        alpha=0.72,
        edgecolors='white',
        linewidths=1.2,
        label=region,
        zorder=3
    )

# Trend line
all_gdp = np.linspace(1000, 60000, 200)
trend   = 50 + (all_gdp / 60000) * 35
ax.plot(all_gdp, trend, '--', color='#aaaaaa',
        linewidth=1.5, alpha=0.7, label='Trend line', zorder=2)

# Annotations
ax.text(0.02, 0.97,
        'Bubble size = Population (millions)',
        transform=ax.transAxes, fontsize=9,
        color='#666', va='top',
        bbox=dict(boxstyle='round,pad=0.4',
                  facecolor='white', edgecolor='#ddd', alpha=0.9))

ax.set_xlabel('GDP per Capita (USD)', fontsize=12)
ax.set_ylabel('Life Expectancy (years)', fontsize=12)
ax.set_title('GDP vs Life Expectancy by Region (Simulated)', pad=15)
ax.legend(title='Region', framealpha=0.9, edgecolor='#ddd')
ax.grid(linestyle='--', alpha=0.3, color='#aaaaaa')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

plt.tight_layout()
plt.show()

# TRY: Change the formula for life_exp to see different correlation shapes
# TRY: Set alpha=1.0 vs alpha=0.2 to see the effect on overlapping dots
````

### When to use a Histogram?

A histogram answers: **"How is my data distributed?"**

It groups continuous data into **bins** and shows how many values fall in each bin. Think of it as a bar chart for a single numerical column.

**What to look for:**
- **Bell curve (normal)** — data clusters around the middle
- **Skewed right** — most values are low, a few are very high (e.g. income)
- **Bimodal (two peaks)** — two distinct groups in your data
- **Uniform** — all values equally likely

````python
# ============================================================
# HISTOGRAM — 4 Distribution Shapes
# ============================================================
np.random.seed(21)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.patch.set_facecolor('#fafbff')
fig.suptitle('Common Data Distribution Shapes', fontsize=16,
             fontweight='bold', y=1.01)

datasets = [
    (np.random.normal(50, 10, 1000), 'Normal Distribution',
     '#4361ee', 'Exam scores, heights, measurement errors'),
    (np.random.exponential(10, 1000), 'Right-Skewed Distribution',
     '#f72585', 'Incomes, city populations, response times'),
    (np.concatenate([np.random.normal(30, 5, 500),
                     np.random.normal(70, 5, 500)]),
     'Bimodal Distribution',
     '#06d6a0', 'Mixed groups, e.g. male + female heights'),
    (np.random.uniform(0, 100, 1000), 'Uniform Distribution',
     '#ffd166', 'Random numbers, dice rolls'),
]

for ax, (data, title, color, example) in zip(axes.flat, datasets):
    ax.set_facecolor('#fafbff')
    ax.hist(data, bins=35, color=color, alpha=0.80,
            edgecolor='white', linewidth=0.6)

    # Mean line
    mean_val = np.mean(data)
    ax.axvline(mean_val, color='#333', linestyle='--',
               linewidth=1.5, label=f'Mean = {mean_val:.1f}')

    ax.set_title(title, pad=10)
    ax.set_xlabel('Value')
    ax.set_ylabel('Count')
    ax.legend(fontsize=9, edgecolor='#ddd')
    ax.text(0.98, 0.95, f'e.g. {example}',
            transform=ax.transAxes, fontsize=7.5,
            ha='right', va='top', color='#888',
            style='italic')
    ax.grid(axis='y', linestyle='--', alpha=0.3, color='#aaa')

plt.tight_layout()
plt.show()

# TRY: Change bins=35 to bins=5 (too few) or bins=200 (too many)
# TRY: Replace np.random.normal with np.random.lognormal
````

### When to use a Pie Chart?

Pie charts show **part-to-whole** relationships — how each category contributes to a total (100%).

> **Honest advice:** Most data scientists prefer **bar charts over pie charts** for comparison accuracy. But pie charts win when:
> - You have **5 or fewer** categories
> - You want to highlight **one dominant slice** (e.g. "50% of budget goes to X")
> - You need a **visually appealing infographic**

**The Donut chart** = pie chart with a hole in the middle. It's trendier and lets you put a summary number in the center!

````python
# ============================================================
# PIE + DONUT CHART
# ============================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 6))
fig.patch.set_facecolor('#fafbff')

labels  = ['Housing', 'Food', 'Transport', 'Entertainment', 'Savings']
sizes   = [35, 25, 15, 12, 13]
colors  = ['#f72585', '#4361ee', '#06d6a0', '#ffd166', '#4cc9f0']
explode = (0.05, 0.05, 0.05, 0.05, 0.08)  # Savings slightly exploded

# ── LEFT: Classic Pie
ax1.set_facecolor('#fafbff')
wedges, texts, autotexts = ax1.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.82,
    wedgeprops=dict(edgecolor='white', linewidth=2)
)
for text in autotexts:
    text.set_fontsize(10)
    text.set_fontweight('bold')
    text.set_color('white')
ax1.set_title('Monthly Budget Breakdown', pad=15)

# ── RIGHT: Donut Chart
ax2.set_facecolor('#fafbff')
wedges2, texts2, autotexts2 = ax2.pie(
    sizes,
    colors=colors,
    autopct='%1.1f%%',
    startangle=140,
    pctdistance=0.82,
    wedgeprops=dict(edgecolor='white', linewidth=2, width=0.55)  # width < 1 = donut!
)
for text in autotexts2:
    text.set_fontsize(10)
    text.set_fontweight('bold')
    text.set_color('white')

# Center text inside donut
ax2.text(0, 0.08, 'Total', ha='center', fontsize=13,
         color='#555', fontweight='bold')
ax2.text(0, -0.18, '$4,200', ha='center', fontsize=18,
         color='#222', fontweight='900')

# Legend outside
ax2.legend(wedges2, labels, loc='lower center',
           bbox_to_anchor=(0.5, -0.12), ncol=3,
           fontsize=9, framealpha=0.9, edgecolor='#ddd')
ax2.set_title('Monthly Budget (Donut Style)', pad=15)

plt.tight_layout()
plt.show()

# TRY: Change explode values to highlight a different slice
# TRY: Swap 'width=0.55' to 'width=0.3' for a thinner ring
````

**When to choose Seaborn:**

````python
# ============================================================
# HEATMAP — Correlation Matrix
# ============================================================
# A heatmap uses COLOR INTENSITY to show values in a grid.
# Most common use: correlation matrix — which variables are related?

np.random.seed(0)

# Simulate a dataset with correlated features
n = 200
age     = np.random.randint(22, 65, n)
salary  = age * 900 + np.random.normal(0, 8000, n)
exp     = (age - 22) * 0.8 + np.random.normal(0, 3, n)
edu     = np.random.randint(1, 5, n).astype(float)
score   = 0.3*salary/1000 + 0.4*exp + 0.2*edu*5 + np.random.normal(0, 5, n)
hours   = np.random.normal(42, 8, n)

df = pd.DataFrame({
    'Age': age, 'Salary': salary, 'Experience': exp,
    'Education': edu, 'Performance': score, 'Hours/Week': hours
})

corr = df.corr()

fig, ax = plt.subplots(figsize=(8, 6))
fig.patch.set_facecolor('#fafbff')

mask = np.triu(np.ones_like(corr, dtype=bool), k=1)  # Show only lower triangle

sns.heatmap(
    corr,
    mask=mask,
    annot=True,          # Show numbers inside cells
    fmt='.2f',           # 2 decimal places
    cmap='RdYlGn',       # Red = negative, Green = positive
    center=0,
    vmin=-1, vmax=1,
    square=True,
    linewidths=2,
    linecolor='white',
    annot_kws={'size': 11, 'weight': 'bold'},
    ax=ax
)

ax.set_title('Correlation Matrix — Employee Dataset', pad=15, fontsize=14)
ax.tick_params(axis='x', rotation=30)
ax.tick_params(axis='y', rotation=0)

plt.tight_layout()
plt.show()

print("Reading a correlation matrix:")
print("  +1.0 = perfect positive correlation")
print("   0.0 = no relationship")
print("  -1.0 = perfect negative correlation")
print(f"\nStrongest correlation: Age <-> Experience = {corr.loc['Age','Experience']:.2f}")

# TRY: Change cmap='RdYlGn' to cmap='coolwarm' or cmap='Blues'
````

````python
# ============================================================
# BOX PLOT + VIOLIN PLOT side by side
# ============================================================
# Both show distribution statistics — but differently!

np.random.seed(5)
departments = ['Engineering', 'Marketing', 'Sales', 'Design', 'HR']
salaries = [
    np.random.normal(95, 15, 80),
    np.random.normal(72, 12, 60),
    np.random.normal(68, 20, 70),
    np.random.normal(78, 10, 50),
    np.random.normal(62, 8,  45),
]

# Build a tidy DataFrame
records = []
for dept, sal in zip(departments, salaries):
    for s in sal:
        records.append({'Department': dept, 'Salary': s})
df_sal = pd.DataFrame(records)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.patch.set_facecolor('#fafbff')
fig.suptitle('Salary Distribution by Department ($K)', fontsize=15,
             fontweight='bold', y=1.02)

dept_colors = ['#f72585','#4361ee','#06d6a0','#ffd166','#4cc9f0']

# ── LEFT: Box Plot
sns.boxplot(
    data=df_sal, x='Department', y='Salary',
    palette=dept_colors,
    linewidth=1.5,
    flierprops=dict(marker='o', markersize=4, alpha=0.5),
    ax=ax1
)
ax1.set_facecolor('#fafbff')
ax1.set_title('Box Plot\n(Median, IQR, Outliers)', pad=10)
ax1.set_xlabel('')
ax1.set_ylabel('Salary ($K)')
ax1.tick_params(axis='x', rotation=20)
ax1.grid(axis='y', linestyle='--', alpha=0.35)

# ── RIGHT: Violin Plot
sns.violinplot(
    data=df_sal, x='Department', y='Salary',
    palette=dept_colors,
    inner='box',       # Show mini box inside
    linewidth=1.2,
    ax=ax2
)
ax2.set_facecolor('#fafbff')
ax2.set_title('Violin Plot\n(Distribution Shape + Stats)', pad=10)
ax2.set_xlabel('')
ax2.set_ylabel('Salary ($K)')
ax2.tick_params(axis='x', rotation=20)
ax2.grid(axis='y', linestyle='--', alpha=0.35)

plt.tight_layout()
plt.show()

print("Box Plot reads:  [whisker — Q1 — MEDIAN — Q3 — whisker]  + outlier dots")
print("Violin Plot adds: the WIDTH shows how many data points are at each value")

# TRY: Change inner='box' to inner='point' or inner='stick'
````

### Why Plotly?

Matplotlib and Seaborn produce **static images**. Plotly produces **interactive HTML charts** you can:
- **Hover** over data points for tooltips
- **Zoom** and **pan**
- **Click legend** items to hide/show series
- **Export** as HTML for websites

> Perfect for dashboards, web apps, and presentations where your audience can explore the data themselves!

### Combining Multiple Charts into a Dashboard

Real-world reports rarely show one chart. You'll often need a **multi-panel dashboard** — multiple charts arranged on a single figure.

The key tool is `plt.subplot_mosaic()` (modern) or `plt.subplots(rows, cols)` (classic).

```python
# Mosaic layout — define with text!
fig, axes = plt.subplot_mosaic("""
    AABB
    CCDD
""")  
# A and B span the top row, C and D span the bottom
```

````python
# ============================================================
# MINI DASHBOARD — Sales Report Layout
# ============================================================
np.random.seed(99)

fig, axes = plt.subplot_mosaic("""
    AABC
    AADD
""", figsize=(15, 9))

fig.patch.set_facecolor('#f0f4ff')
for ax in axes.values():
    ax.set_facecolor('#fafbff')

# ── A: Main Line Chart (big left panel)
months  = ['Jan','Feb','Mar','Apr','May','Jun',
           'Jul','Aug','Sep','Oct','Nov','Dec']
revenue = [42,48,45,55,60,72,68,75,82,79,88,96]
target  = [50,52,55,58,62,65,68,70,75,78,82,85]

axes['A'].plot(months, revenue, 'o-', color='#f72585',
               linewidth=2.5, markersize=7, label='Actual')
axes['A'].plot(months, target, '--s', color='#4361ee',
               linewidth=2, markersize=6, alpha=0.7, label='Target')
axes['A'].fill_between(range(12), revenue, target,
                        where=[r>t for r,t in zip(revenue, target)],
                        alpha=0.15, color='#06d6a0', label='Above target')
axes['A'].fill_between(range(12), revenue, target,
                        where=[r<=t for r,t in zip(revenue, target)],
                        alpha=0.15, color='#f72585', label='Below target')
axes['A'].set_xticks(range(12))
axes['A'].set_xticklabels(months, rotation=15)
axes['A'].set_title('Revenue vs Target ($K)', fontsize=13)
axes['A'].legend(fontsize=9, edgecolor='#ddd')
axes['A'].grid(linestyle='--', alpha=0.3)

# ── B: KPI Card (top right)
axes['B'].axis('off')
axes['B'].add_patch(mpatches.FancyBboxPatch(
    (0.05, 0.05), 0.9, 0.9,
    boxstyle='round,pad=0.02',
    facecolor='#f72585', edgecolor='white', linewidth=3,
    transform=axes['B'].transAxes, zorder=0
))
axes['B'].text(0.5, 0.65, '$796K', ha='center', va='center',
               fontsize=26, fontweight='900', color='white',
               transform=axes['B'].transAxes)
axes['B'].text(0.5, 0.35, 'Total Revenue', ha='center',
               fontsize=12, color='#f0e6ff',
               transform=axes['B'].transAxes)
axes['B'].text(0.5, 0.18, 'YTD 2024', ha='center',
               fontsize=9, color='#ddd0f5',
               transform=axes['B'].transAxes)

# ── C: Donut (top far right)
cat_labels = ['Online', 'Retail', 'B2B']
cat_vals   = [55, 28, 17]
cat_colors = ['#4361ee', '#06d6a0', '#ffd166']
axes['C'].pie(cat_vals, labels=cat_labels, colors=cat_colors,
              autopct='%1.0f%%', startangle=90,
              pctdistance=0.78,
              wedgeprops=dict(width=0.5, edgecolor='white', linewidth=2))
axes['C'].set_title('Sales Channel Mix', fontsize=11)

# ── D: Bar chart (bottom right)
regions  = ['North', 'South', 'East', 'West']
reg_vals = [230, 185, 210, 171]
reg_cols = ['#f72585','#4361ee','#06d6a0','#ffd166']
bars = axes['D'].bar(regions, reg_vals, color=reg_cols,
                     edgecolor='white', linewidth=1.5, width=0.55)
for bar, val in zip(bars, reg_vals):
    axes['D'].text(bar.get_x() + bar.get_width()/2,
                   val + 3, f'${val}K',
                   ha='center', fontsize=9, fontweight='bold')
axes['D'].set_title('Revenue by Region ($K)', fontsize=11)
axes['D'].grid(axis='y', linestyle='--', alpha=0.3)
axes['D'].set_ylim(0, 260)

fig.suptitle('Sales Performance Dashboard — 2024',
             fontsize=17, fontweight='900', y=1.01, color='#1a1a2e')
plt.tight_layout(pad=2)
plt.show()
````
